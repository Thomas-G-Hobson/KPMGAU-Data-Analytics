## Model Building and Evaluating

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
# Spilt data into training set and testing set
X_train, X_test, y_train, y_test = train_test_split(df_num, df_num["past_3_years_bike_related_purchases"], test_size=0.25)
```

```python
# Build the decision tree model.
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
```

```python
# Test the model on the test set.
y_pred = clf.predict(X_test)
```

```python
# Check the necessary metrics.
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
feature_importances = clf.feature_importances_

# Label the features
feature_names = list(X_train.columns)

# Print the feature importances with labels
print(pd.DataFrame(
    data={"Feature": feature_names, "Importance": feature_importances},
    index=None,
    columns=["Feature", "Importance"],
))
```
